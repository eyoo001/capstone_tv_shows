```python
viewership = []
table = soup.find('table', {'class':'wikitable plainrowheaders wikiepisodetable'})
row = soup.find('tr')
for row in table.find_all('tr', {'class': 'vevent'}):
    rating = {}
    rating['name'] = row.text
#    rating['slug'] = row['tr'].split('/')[-1]
    viewership.append(rating)
```
